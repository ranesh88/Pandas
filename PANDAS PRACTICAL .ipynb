### Merge, concatenate and join - Pandas

```python
import pandas as pd
data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
        "Name":["Ravi","Lisa","John","Tulika","Gaurav","Subho"],
        "Age":[25,36,26,34,31,29]} 


data2 ={"Emp ID":["E01","E02","E03","E04","E05","E06"],
       "Department":["Finance","HR","Analytics","HR","IT","Marketing"],
       "Salary":[35000,42000,60000,25000,55000,32000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2,on = 'Emp ID',how = 'inner')) ## default
```

```python
data1 = {"Emp ID":["E01","E02","E03","E04","E05","E06"],
        "Name":["Ravi","Lisa","John","Tulika","Gaurav","Subho"],
        "Age":[25,36,26,34,31,29]} 


data2 ={"Emp ID":["E01","E07","E03","E04","E08","E06"],
       "Department":["Finance","HR","Analytics","HR","IT","Marketing"],
       "Salary":[35000,42000,60000,25000,55000,32000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2,on = 'Emp ID')) 
```

```python
print(pd.merge(left = df1,right = df2,on = 'Emp ID',how = 'left')) ## left merge
```

```python
print(pd.merge(left = df1,right = df2,on = 'Emp ID',how = 'right')) ## right merge
```

### Pandas-Compare DataFrames

```python
dict = {"Fruits":["Mango","Apple","Banana","Guava"],
       "Price":[150,100,25,35],
       "Quantity":[15,20,25,14]}

df1 = pd.DataFrame(dict)

print(df1)

df2 = df1.copy()
print(df2)
```

```python
df2.loc[0,"Price"]=175
df2.loc[1,"Price"]=152
df2.loc[2,"Price"]=50
df2.loc[3,"Price"]=32

df2.loc[0,"Quantity"]=25
df2.loc[1,"Quantity"]=28
df2.loc[2,"Quantity"]=32
df2.loc[3,"Quantity"]=35

print(df2)
```

```python
print(df1.compare(df2))
```

```python
print(df1.compare(df2,align_axis=0))
```

### Pivoting and Melting Dataframes

```python
# Use pivot_table to handle multiple values
dict = {"Keys":["K1","K2","K1","K2"],
       "Names":["Peter","John","Ben","David"],
       "Houses":["Red","Green","Blue","Red"],
       "Grades":["8th","9th","4th","7th"]}

df = pd.DataFrame(dict)
print(df)

pivoted_df = df.pivot_table(index="Keys", columns="Names", values=["Houses", "Grades"],aggfunc=lambda x: x)
print(pivoted_df)
```

```python
dict = {"Names":["Peter","John","Ben","David"],
       "Houses":["Red","Green","Blue","Red"],
       "Grades":["8th","9th","4th","7th"]}

df = pd.DataFrame(dict)
print(df)
```

```python
print(pd.melt(df,id_vars = ["Names"],value_vars = ["Houses","Grades"],var_name ='Houses & Grades',value_name='values'))
```
